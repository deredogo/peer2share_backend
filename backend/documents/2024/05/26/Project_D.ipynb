```python
## A. Build a baseline model
```

```python
import pandas as pd
import numpy as np
import keras
#import sklearn
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./concrete_data.csv')
data.head(15)
```

```python
X = data.drop(columns=['Strength'])
y = data['Strength']
```

```python
data_columns = data.columns
target = data['Strength']
target.head(10)
```

```python
predictors = data.iloc[:, :-1]
predictors.head(10)
```

```python
n_cols = predictors.shape[1]
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
      
```

```python
model = regression_model()
```

```python
mse_list = []
```

```python

for cycle in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
    res = model.fit(X_train, y_train, epochs=50, verbose=0, validation_data=(X_test, y_test))
    mean_squared_error = res.history['val_loss'][-1]
    mse_list.append(mean_squared_error)
    print('{}: Mean squared error {}'.format(cycle+1, mean_squared_error))
```

```python
mean_sqmse_A = np.mean(mse_list)
std_sqmse_A  = np.std(mse_list)

#generate a data frame 
results = pd.DataFrame.from_dict({"Part": ["A"],"Mean of Mean Squared Errors": [mean_sqmse_A], "Standard deviation of Mean Squared Errors": [std_sqmse_A]})
results
```

```python
# Save the results and display
results = pd.DataFrame(mse_list, columns=['Mean Squared Error'])
results.head(10)
```

```python
## Part B: Normalize the data 
```

```python
predictors_n = (predictors - predictors.mean())/predictors.std()
predictors_n.head(10)
```

```python
#Building Model
cols = predictors_n.shape[1]
def second_regression_model():
    modelN = Sequential()
    modelN.add(Dense(10, activation='relu', input_shape=(cols,)))
    modelN.add(Dense(1))
    
    modelN.compile(optimizer='adam', loss='mean_squared_error')
    return modelN

modelN = second_regression_model()
```

```python
mse_list_b= []
for cycle in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_n, target, test_size=0.3)
    res = modelN.fit(X_train, y_train, epochs=50, verbose=0, validation_data=(X_test, y_test))
    mean_squared_error = res.history['val_loss'][-1]
    mse_list_b.append(mean_squared_error)
    print('{}: Mean Squared Error {}'.format(cycle+1, mean_squared_error))
```

```python
mean_sqmse_B = np.mean(mse_list_b)
std_sqmse_B  = np.std(mse_list_b)

#generate a data frame 
results.loc[results.shape[0]] = ["B",mean_sqmse_B,std_sqmse_B]
results
```

```python
# Save the results and display
results = pd.DataFrame(mse_list, columns=['Mean Squared Error'])
results.head(10)
```

```python
## C. Increate the number of epochs
```

```python
def third_regression_model():
    modelE = Sequential()
    modelE.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    modelE.add(Dense(1))
    
    modelE.compile(optimizer='adam', loss='mean_squared_error')
    return modelE

modelE = third_regression_model() 
```

```python
mse_list_c = []
for cycle in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_n, target, test_size=0.3)
    res = modelE.fit(X_train, y_train, epochs=100, verbose=0, validation_data=(X_test, y_test))
    mean_squared_error = res.history['val_loss'][-1]
    mse_list_c.append(mean_squared_error)
    print('{}: Mean Squared Error {}'.format(cycle+1, mean_squared_error))
```

```python
mean_sqmse_C = np.mean(mse_list_c)
std_sqmse_C  = np.std(mse_list_c)

results.loc[results.shape[0]] = ["C",mean_sqmse_C,std_sqmse_C]
results
```

```python
# We repeated Part B but this time we used 100 epochs for training.
```

```python
## D. Increase the number of hidden layers
```

```python
def fourth_regression_model():
    modelNN = Sequential()
    modelNN.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    modelNN.add(Dense(10, activation='relu'))
    modelNN.add(Dense(10, activation='relu'))
    modelNN.add(Dense(1))
    
    modelNN.compile(optimizer='adam', loss='mean_squared_error')
    return modelNN
```

```python
modelNN = fourth_regression_model()
```

```python
mse_list_d = []
for cycle in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_n, target, test_size=0.3)
    res = modelNN.fit(X_train, y_train, epochs=50, verbose=0, validation_data=(X_test, y_test))
    mean_squared_error = res.history['val_loss'][-1]
    mse_list_d.append(mean_squared_error)
    print('{}: Mean Squared Error {}'.format(cycle+1, mean_squared_error))
```

```python
mean_sqmse_D = np.mean(mse_list_d)
std_sqmse_D  = np.std(mse_list_d)

results.loc[results.shape[0]] = ["D",mean_sqmse_D,std_sqmse_D]
results
```
